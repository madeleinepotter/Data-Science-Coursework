# Investigating the Effectiveness of Physical Building and Socio-Demographic Factors in Predicting CO₂ Emissions From Gas Consumption In London's Residential Sector

## 1. Introduction

In the face of climate change, London has set a target of becoming a Net Zero Carbon city by 2030. To achieve this, the residential sector which in the UK generated 17% of total CO₂ emissions in 2022 (Department for Energy Security and Net Zero, 2023) will have to be addressed. A significant proportion of these emissions are linked to household energy use, specifically originating from the combustion of natural gas for heating and cooking (CCC,2020). Although progress in emissions reductions is relatively fast, London's residential sector remains a key area of concern as reducing these emissions is a significant challenge to undertake. This is mainly because factors affecting emitted emission levels are linked to both the physical characteristics of the building itself and the behaviours of the residents within them. 

Understanding the complex nature of these factors in the residential sector can help the city achieve its Net Zero goal, while also preventing any possible harm to vulnerable communities (Büchs et al., 2011).
.


## 2. Research Question

Therefore, to better understand factors affecting CO₂ emissions in the residential sector to allow for more directed action, the following research questions are posed:

**1. How effectively can physical building factors and socio-demographic factors predict CO2 emissions from gas consumption in London's residential sector?**

**2. Which of these factors influence the prediction performance the most?**

This study focuses on gas consumption as it is a significant contributor to CO₂.

## 3. Literature Review 

Despite a significant proportion of CO₂ emissions in the UK and specifically London, originates from the residential sector, few studies have assessed contributing factors. Schuster et al., (2023) assessed the impact of socio-demographic and housing characteristics on CO₂ emissions at a regional UK level using a multivariate regression analysis. It found stark differences in CO₂ emissions across regional households, with living space size being the biggest determinant. 

Although most studies focus on factors influencing gas and electricity consumption to understand household energy efficiency, these studies also provide insights into factors driving CO₂ emissions as energy consumption directly relates to emissions (Drummond and Ekins, 2016). Recentlys, machine learning (ML) method such as linear regression, artificial neural networks, support vector machines and ensemble learning approaches (such as random forest) have been increasingly used to explore this topic. Studies state that suchh methods reveal more information while overcoming the shortfalls of traditional research methods such as conducting surveys (Chen et al., 2022). While ANN and SVM show high model performance, due to their lack of interpretabilit,y other methods are often used in this area of study (Ma and Cheng, 2016). For example, random forest is one of the most widely used ML methods due to its ease of use and interpretability (Ahmed et al., 2019). Another commoly used algorithmd is gradient boosting, which was found to perform better than random forest models in a study thao predicedt commercial building energy consumption (Robinson et al., 2017).
The literature indicates both h building characteristics and socio-demographic characteristicd aimportant s determinants of CO₂ emissions. Dwelling size is often associated with higher energy consumption for heating, as is building age, which is likely due to the lower energy efficiency of older buildings (Namazkhan et al., 2020). With the UK having some of the oldest housing stock in Europe and therefore also some of the least energy efficient (Broad et al., 2020) this is often a significant area of focus and retrofitting a favourable solution (London Councils, n.d). However, socio-demographic characteristics of the people who reside in buildings also contribute, with income, age, and education level impacting energy consumption (Heubner et al., 2015). Income has been found to be a significant driver, likely since households with higher incomes reside in larger dwellings and are less constricted by their budget to use energy (Hara et al., 2015)Therefore, addressing both areas is important for addressing CO₂ emissions in the residential sector.

. Again, while these studies focus on household energy consumption, they can still give useful insightinto drivers ofof household CO₂ emissio.. 


## 4. Methodology

To assess how well physical building and socio-demographic factors can predict CO2 emissions from gas consumption in London’s residential sector, two ensemble regression methods will be used: Random Forest and XG Boost. 

The data was collected from the Place Based Carbon Calculator and the London Data Store, where data is stored by LSOAs in London. As LSOAs are geographic units of observation, the modelling approach had to account for spatial dependence in the data. To do this, the global Moran’s I statistic was calculated to determine if spatial autocorrelation existed in the dependent variable. As it was present in the data, a spatial lag variable was included in the model to improve its accuracy, as suggested by Liu et al., (2022). 

Independent variables (factors influencing CO₂ emissions from household gas consumption) were chosen based on the research mentioned in the literature review and once the data was cleaned and spatial dependence was accounted for, this was split into training (75%) and testing data (25%). K-fold cross validation was used for both models, with the number of folds set to 5 to not be too computationally expensive. The R squared and root-mean-square error (RMSE) were then used to assess the performance of both models. Finally, a permutation feature importance (PFI) was done for both models and compared to see which variable influenced the prediction performance the most.

### 5.6 Checking for Spatial Autocorrelation

Previous research demonstrated CO₂ emissions from households are often spatially autocorrelated (Liu et al., 2017;Long et al., 2020) Failing to address spatial autocorrelation can result in over or underpredicted CO₂ emission levels from household gas consumption (Hu et al., 2022), especially if the variable is highly spatially dependent. 

Therefore, to determine if <code>gas_percap_2011</code> is spatially autocorrelated, the Moran's I statistic is calculated and plotted.

```python
# removing nan values to not produce errors for the spatial weights calculations

joined_gdf = joined_gdf.dropna(subset=['gas_percap_2011'])
```

```python
# calculating spatial weights matrix using centroids of the polygons (each LSOA) so we can assess the Moran's I statistics
# and see if spatial autocorrelation present and needed to be handled and included in analysis 

# Calculate centroids of polygons (LSOAs)
joined_gdf['centroid'] = joined_gdf.geometry.centroid
joined_gdf['centroid_x'] = joined_gdf['centroid'].x
joined_gdf['centroid_y'] = joined_gdf['centroid'].y

# Create a GeoDataFrame from the centroids
centroids_gdf = gpd.GeoDataFrame(joined_gdf, geometry=joined_gdf['centroid'])

# Create spatial weights using the centroids
w_knn = weights.KNN.from_dataframe(centroids_gdf, k=5)

# Ensure weights are row-standardized
w_knn.transform = 'R'

# z-score normalisation to create standardised values for gas_percap_2011
joined_gdf['gas_percap_2011_std'] = (joined_gdf['gas_percap_2011'] - joined_gdf['gas_percap_2011'].mean()) / joined_gdf['gas_percap_2011'].std()

# Calculating spatial lag of the standardized values
joined_gdf['gas_percap_2011_lag_std'] = weights.spatial_lag.lag_spatial(w_knn, joined_gdf['gas_percap_2011_std'])

# Calculating Moran's I using standardized variable
moran_i_std = Moran(joined_gdf['gas_percap_2011_std'], w_knn)

print(f"Standardized Moran's I: {moran_i_std.I} with p-value {moran_i_std.p_sim}")
```

```python
# Plotting Moran's Scatterplot

fig, ax = plt.subplots(figsize=(10, 10))

# Scatter plot
sns.regplot(x='gas_percap_2011_std', y='gas_percap_2011_lag_std', data=joined_gdf,
            ci=None, scatter_kws={'color': 'b'}, line_kws={'color': 'r'})

# Draw lines at x=0 and y=0 to divide the plot into quadrants
plt.axhline(0, color='grey', linestyle='--')  # Horizontal line at y=0
plt.axvline(0, color='grey', linestyle='--')  # Vertical line at x=0

# Annotate quadrants. 
# HH: High High
# HL: High Low
# LL: Low Low
# LH: Low High

ax.text(0.05, 0.95, 'HH', transform=ax.transAxes, verticalalignment='top', horizontalalignment='right')
ax.text(0.05, 0.05, 'HL', transform=ax.transAxes, verticalalignment='bottom', horizontalalignment='right')
ax.text(0.95, 0.05, 'LL', transform=ax.transAxes, verticalalignment='bottom', horizontalalignment='left')
ax.text(0.95, 0.95, 'LH', transform=ax.transAxes, verticalalignment='top', horizontalalignment='left')

# Set titles and labels
ax.set_title("Moran's I Scatterplot")
ax.set_xlabel("Standardized CO₂ Emissions Per Capita from Household Gas Consumption")
ax.set_ylabel("Spatial Lag of Standardized CO₂ Emissions Per Capita from Household Gas Consumption")

plt.show()
```

The Moran's I value of 0.548 (with a p-value < 0.05) and the Moran's I Scatterplot indicate a slight positive spatial autocorrelation is present in the dependent variable. Under such circumstances, including spatial lag into machine learning models often improves the accuracy of estimations and predictive performance (Liu et al., 2022). 

Therefore, <code>gas_percap_2011_lag_std</code> will be included in the model dataset.

### 6.1 Random Forest Model

### 6.1.1 Testing and Training Split

```python
final_gdf = joined_gdf.copy()
```

```python
# dropping categorical variables that relate to the LSOA values that do not need to be in the model

final_gdf = final_gdf.drop(['LSOA11CD', 'LSOA11NM_left', 'geometry', 'LSOA11NM_right', 'SOAC11NM', 'LAD17CD', 'centroid', 'centroid_x', 
               'centroid_y', 'Names'], axis=1)
```

```python
final_gdf.columns
```

```python
# default split is 75% training, 25% testing so no need to set specific values
random_state_split = 100
train_x, test_x, train_y, test_y = train_test_split(final_gdf.drop(['gas_percap_2011','gas_percap_2011_std'], axis = 1), final_gdf.gas_percap_2011, random_state=random_state_split)
```

```python
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)

# making sure train_x and train_y index are the same

print(train_x.index.identical(train_y.index))
print(test_x.index.identical(test_y.index))
```

### 6.1.2 Building Random Forest

```python
# to figure out how long it takes to run
start_time = time.time()
elapsed_time = time.time() - start_time

# values of max_depth and min_samples_split
hyperparameters = {'max_depth':[10,20,30,40,50], 'min_samples_split':[2,4,6,8,10]}

randomState_dt = 10000
rf = RandomForestRegressor(random_state=randomState_dt)

# cv=5 by default, which means 5-fold cross-validation
clf = GridSearchCV(rf, hyperparameters)

clf.fit(train_x, train_y)

# wprint out the best parameter and score
print ("The best parameter value is: ")
print (clf.best_params_)
print ("The best score is: ")
print (clf.best_score_)

print(f"Elapsed time to compute best fit parameter values and score: {round((elapsed_time)/60,2)} minutes")
```

```python
# fit final random forest model
start_time = time.time()
elapsed_time = time.time() - start_time

rf_final = RandomForestRegressor(max_depth=clf.best_params_['max_depth'], min_samples_split=clf.best_params_['min_samples_split'], random_state=randomState_dt)
rf_final.fit(train_x, train_y)

print(f"Elapsed time to compute the model: {round((elapsed_time)/60,2)} minutes")
```

### R Squared Value

```python
print("R2 on the training data for Random Forest:")
rf_rsq_training = rf_final.score(X=train_x, y=train_y)
print(rf_rsq_training)

print("R2 on the testing data for Random Forest:")
rf_rsq_testing = rf_final.score(X=test_x, y=test_y)
print(rf_rsq_testing)
```

### RMSE

```python
print("RMSE on the training data for Random Forest:")
rf_rmse_training = mean_squared_error(train_y, rf_final.predict(train_x), squared=False)
print(rf_rmse_training)

print("RMSE on the testing data for Random Forest:")
rf_rmse_testing = mean_squared_error(test_y, rf_final.predict(test_x), squared=False)
print(rf_rmse_testing)
```

### 6.1.3 Permutation Feature Importance 

```python
imp = rfpimp.importances(rf_final, test_x, test_y)
print(imp)
viz = rfpimp.plot_importances(imp)
viz.view()
```

The PFI shows average floor area is the most influential with a performance value of 0.270. After that, houses owned outright and the spatial lag variable for CO₂ emissions are the most influential, however significantly less so with their importance values being 0.0979 and 0.0458 respectively.
